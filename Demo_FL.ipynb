## Parameter Setting

```python
args = {
    "batch_size": 64,
    "test_batch_size": 128,
    "epochs": 5,
    "lr": 0.01,
    "gamma": 0.7,
    "no_cuda": True,
    "dry_run": False,
    "seed": 42, # the meaning of life
    "log_interval": 10,
    "save_model": True,
}
```

## Validation data

```python
df_all = pd.read_csv('data/test.csv')
```

```python
features = df_all.columns[2:]
target = 'label'
```

```python
df_all.sample(5)
```

```python
X = torch.tensor(df_all[features].values, dtype=torch.float32)
y = torch.tensor(df_all[target].values, dtype=torch.long)
```

```python
valid_ds = TensorDataset(X, y)
valid_dl = DataLoader(valid_ds, batch_size=args["test_batch_size"])
```

```python
test_data_length = len(valid_dl.dataset)
print(test_data_length)
```

## Model

```python
local_model = Basic_CNN(torch,len(features))
```

## Remote work

### Wrap it into one

```python
all_client = [duet,duet2]
```

### Training

```python
import time

#args["dry_run"] = True  # comment to do a full train
    
print("Starting Training")
for epoch in range(1, args["epochs"] + 1):
    epoch_start = time.time()
    print(f"Epoch: {epoch}")
    
    # send model
    all_model = [local_model.send(client) for client in all_client]
    all_ref = [client.torch for client in all_client]
    
    # check device capability
    cuda_capable = [check_cuda_capable(ref,no_cuda=args["no_cuda"], seed=args["seed"]) for ref in all_ref]

    for index,client_model in enumerate(all_model):
        if cuda_capable[index]:
            client_model.cuda(device)
        else:
            client_model.cpu()
    
    # get data length at this round
    all_data_length = [get_train_length(client, 2) for client in all_client]
    all_trainloader = [get_dataloader(client, 2, args["batch_size"])for client in all_client]
    
    # remote training on model with remote_torch
    for index,client_model in enumerate(all_model):
        print(f"Client : {index+1}")
        print(f"Data available: {all_data_length[index]}")
        train(client_model, 
              all_client[index], 
              all_trainloader[index], 
              epoch, 
              args, 
              all_data_length[index], 
              batch_size=args["batch_size"],
              log_interval=args["log_interval"],
              dry_run = args["dry_run"]
             )
    
    # update model
    update_gradient(local_model, all_model, all_data_length)
    
    # local testing on model with local torch
    test_local(local_model, 
               torch, 
               valid_dl, 
               test_data_length, 
               batch_size = args["test_batch_size"],
               dry_run = args["dry_run"]
              )
    
    epoch_end = time.time()
    print(f"Epoch time: {int(epoch_end - epoch_start)} seconds")
    if args["dry_run"]:
        break
print("Finished Training")
```
